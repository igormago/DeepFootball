```python
import pandas as pd
import numpy as np
import math
import sys
import os
import json
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
from mpl_toolkits.axes_grid1 import AxesGrid

plt.style.use('ggplot')
path_project = "/home/igormago/projects/DeepFootball/"
sys.path.insert(1, path_project)
from conf.configparser import conf

PATH_EXPERIMENTS = os.path.join(conf['path']['experiments'], 'btts_v2')
from sklearn.metrics import accuracy_score
from sklearn.metrics import brier_score_loss
```

```python
basic = ['performance', 'gw', 'basic']
dcw = ['performance', 'gw', 'dcw']
rs = ['performance', 'gw', 'rs']
experiments = [basic, dcw, rs]

CHAMPIONSHIPS =  ['Brazil A', 'Brazil B', 'England A', 'France A', 'Germany A', 'Italy A', 
                 'Netherlands A', 'Portugal A', 'Spain A']

resume_acc = pd.DataFrame()
resume_bl = pd.DataFrame()
for e in experiments:
    
    for c in CHAMPIONSHIPS:
        
        df_path = os.path.join(PATH_EXPERIMENTS, e[0], e[1], e[2], 'predictions', 'test.csv')
        df = pd.read_csv(df_path)

        df = df[(df['m.rg'] > 4) & (df['m.ss.ch.nm'] == c)]

        acc = accuracy_score(df['pred'], df['m.res.bts'])
        bl = brier_score_loss(df['obs.bts.y'], df['pred_prob_y'])
    
        resume_acc.loc[e[2],c] = acc
        resume_bl.loc[e[2],c] = bl
    
    df_path = os.path.join(PATH_EXPERIMENTS, e[0], e[1], e[2], 'predictions', 'test.csv')
    df = pd.read_csv(df_path)

    df = df[(df['m.rg'] > 4)]

    acc = accuracy_score(df['pred'], df['m.res.bts'])
    bl = brier_score_loss(df['obs.bts.y'], df['pred_prob_y'])

    resume_acc.loc[e[2],'Total'] = acc
    resume_bl.loc[e[2],'Total'] = bl
```
